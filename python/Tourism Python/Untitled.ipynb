```python
def pi(it):
    pi = 0
    for n in range(0, it):
        pi += ((-1)**n)/(2*n+1)
    pi*=4
    return pi
```

```python
X = []
PI = []
for i in range(1, 21):
    PI.append(pi(i))
    X.append(str(i))
print(PI)
```

```python
from bokeh.io import push_notebook, output_notebook, show
from bokeh.plotting import figure
output_notebook()

p = figure(x_range = X, title = 'Throwing dices many times')
p.line(X, PI, line_width = 3, color = "blue")
p.line(X, 3.14, line_width = 3, color = "red")

output_notebook()
show(p)
```

```python
import random

#Parametrs of experiment
dices = 4
throws = 150000
#--------------------------------------

#Prepare an array for results
tab = []
results = []

m = dices*6 - dices + 1 # Calculating the size of an array
for i in range(m):
  tab.append(0)
for i in range(m):
  results.append(0)
#print(tab)
#------------------------

#Function that updates the array tab with resluts of throws
def throw(dices):
  result = 0
  for j in range(dices):
    result += (int(6*random.random())+1)    
  tab[result - dices] += 1
    
# repetin throws (trows times)
for i in range(throws):
 throw(dices)

# Showing the results
print("We throw %d with %d dices. The results:" %(throws, dices)) 
print()

for i in range(len(tab)):
  results[i] = round(100*(tab[i]/throws),2)
  print ("%2d  %8d   %5.2f" %(i+dices, tab[i], round(100*(tab[i]/throws),2)))
```

```python
from bokeh.io import push_notebook, output_notebook, show
from bokeh.plotting import figure, output_file
output_notebook()
p = figure(x_range = X, title = 'Throwing dices many times')
p.line(X, results, legend_label="Guass distribution", line_width = 3, color = "blue")

output_notebook()
output_file("chart.html")
show(p)
```
